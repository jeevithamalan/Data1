```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
import plotly.express as px
```

```python
df["Admitted"]=df["Admitted"].replace({"Yes":1,"No":0})
df["Gender"]=df["Gender"].replace({"Female":5,"Male":10})
df
```
